# 1.2. Creation de nouvelle variable 

```python
#Défini une note à chaque categorie
dict_a = {'Ex': 10, 'Gd': 8, 'TA': 6, 'Fa':4, 'Po':2}
dict_b = {'Ex': 10, 'Gd': 8, 'TA': 6, 'Fa':4, 'Po':2, 'NA':0}
dict_c = {'Gd': 8, 'Av': 6, 'Mn':4, 'No':2, 'NA':0}
```

### 1. TOTAL Bath (FulBath + HalfBath) NUM

```python
# TOTAL Bath (FulBath + HalfBath) NUM

df["TotalFullBath"] = df["Bsmt Full Bath"] + df["Full Bath"]
df["TotalHalfBath"] = df["Bsmt Half Bath"] + df["Half Bath"]
df["Total Bath"] = (df["TotalHalfBath"]/2) + df["TotalFullBath"]
```

### 2. Overall (Classement de la qualité, Évalue l'état général de la maison, des materriaux et la finition) NUM


```python
# Overall (Classement de la qualité Évalue l'état général de la maison, des materriaux et la finition) NUM

df["Overall"] = df["Overall Cond"] + df["Overall Qual"]
```

###  3. Exter (Classement de la qualité Évalue l'état général de la maison, des materriaux et la finition) NUM


```python
# Exter (Classement de la qualité Évalue l'état général de l'exterieur, des materiaux et la finition) NUM

#Défini une note à chaque categorie
df["Exter Qual"] = df["Exter Qual"].apply(lambda x: dict_a.get(x))
df["Exter Cond"] = df["Exter Cond"].apply(lambda x: dict_a.get(x))

df["Exter"] = df["Exter Qual"] + df["Exter Cond"]
```

### 4. Garage (Evalue la Qualité et l'etat du garage)

```python
# Garage (Evalue la Qualité et l'etat du garage)

#Défini une note à chaque categorie
df["Garage Cond"] = df["Garage Cond"].apply(lambda x: dict_b.get(x))
df["Garage Qual"] = df["Garage Qual"].apply(lambda x: dict_b.get(x))

df["Garage"] = df["Garage Qual"] + df["Garage Cond"] 
```

### 5. Surface_plus (Prend en compte les m2 des surfaces du garage + sous sol)

```python
# Surface_plus (Prend en compte les m2 des surfaces du garage + sous sol)

df["Surface_sup"] = df["Garage Area"] + df["Total Bsmt SF"] 
```

### 6. Renovation (Nombre d'année depuis la derrnière renovation)

```python
df['Renovation'] = 2022 - df['Year Remod/Add']
```

### 7. Surface_plus_ext (Prend en compte les m2 des surfaces de la veranda + porch fermé)

```python
### Surface_plus_ext (Prend en compte les m2 des surfaces de la veranda + porch fermé)

df["Surface_sup_ext"] = df["Enclosed Porch"] + df["Screen Porch"] 
```

### 8. Bsmt (Condition et Qualité du sous sol + l'exposition du bati)


```python
#Bsmt (Condition et Qualité du sous sol)

df["Bsmt Cond"] = df["Bsmt Cond"].apply(lambda x: dict_b.get(x))
df["Bsmt Qual"] = df["Bsmt Qual"].apply(lambda x: dict_b.get(x))
```

```python
df["Bsmt Exposure"] = df["Bsmt Exposure"].apply(lambda x: dict_c.get(x))
```

```python
df["Bsmt"] = df["Bsmt Cond"] + df["Bsmt Qual"] + df["Bsmt Exposure"]
```
